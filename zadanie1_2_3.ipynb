## 1. Построить график функции y(x)

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Вариант 10

```python
# Определение функций
functions = [
    lambda x: x * np.sin(10* x) **2
]

# Детали функций и их диапазоны х
function_details = [
    ("$x \cdot \sin^{2}(10x)$", [0,2]) 
]

# Построение графиков
plt.figure(figsize=(12, 10))
for i, (func, (title, x_range)) in enumerate(zip(functions, function_details), start=10):
    x_values = np.arange(x_range[0], x_range[1], 0.01 if i != 8 else 0.02) #установка шага для функции 8
    y_values = func(x_values)
    plt.subplot(3, 2, i-5)
    plt.plot(x_values, y_values)
    plt.title(f"Функция {i}: {title}")
    plt.xlabel('x')
    plt.ylabel('y')

plt.tight_layout()
plt.show()
```

## Постройте в одном окне графики двух функции y1(x) и y2(x)

```python
import matplotlib.pyplot as plt
import numpy as np

def func1(x):
    return 10 *x

def func2(x):
    return np.sin(10*x)

# Создание массива значений x
x_values = np.arange(0, 2, 0.01)

# Вычисление значений y для каждой функции в каждой паре
y_values_1 = func1(x_values)
y_values_2 = func2(x_values)


# Построение графиков
# plt.figure(figsize=(15, 10))

# plt.subplot(3, 2, 1)
plt.plot(x_values, y_values_1, label='$y_1(x) = 10x$')
plt.plot(x_values, y_values_2, label='$y_2(x) = \sin(10x)$')
plt.title('Графики функций 1 и 2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()
```

### Используя функцию `subplots()`

```python
def func1(x):
    return 10 * x

def func2(x):
    return np.sin(10 * x)

def func3(x):
    return x * np.sin(10 * x) ** 2

# Создание массива значений x
x_values = np.arange(0, 2, 0.01)

# Вычисление значений y для каждой функции
y_values_1 = func1(x_values)
y_values_2 = func2(x_values)
y_values_3 = func3(x_values)

# Построение графиков в разделенных окнах
fig, axs = plt.subplots(1, 3, figsize=(17, 6))

# График 1
axs[0].plot(x_values, y_values_1, label='$y_1(x) = 10x$')
axs[0].set_title('График функции $y_1(x)$')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].legend()

# График 2
axs[1].plot(x_values, y_values_2, label='$y_2(x) = \sin(10x)$', color='orange')
axs[1].set_title('График функции $y_2(x)$')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].legend()

# График 3
axs[2].plot(x_values, y_values_3, label='$y_3(x) = x \cdot \sin^2(10x)$', color='red')
axs[2].set_title('График функции $y_3(x)$')
axs[2].set_xlabel('x')
axs[2].set_ylabel('y')
axs[2].legend()

plt.tight_layout()
plt.show()
```

### Свободная компоновка с GridSpec

```python
from matplotlib.gridspec import GridSpec

# создания массива значений х
x_values = np.arange(0 , 2, .01)

# вычисление значений у для каждой функции
y_values_1 = func1(x_values)
y_values_2 = func2(x_values)
y_values_3 = func3(x_values)

# построение графиков в разделенный окнах
fig = plt.figure(figsize=(9, 9))

# создание сетки для компоновки графиков 
gs = GridSpec(1, 3, figure=fig) 
gs= fig.add_gridspec(2, 2)


# График 1
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(x_values, y_values_1, label='$y_1(x) = 10x$')
ax1.set_title('График функции $y_1(x)$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()
ax1.grid(True)

# График 2
ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(x_values, y_values_2, label='$y_2(x) = \sin(10x)$', color='darkorange')
ax2.set_title('График функции $y_2(x)$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()
ax2.grid(True)

# График 3
ax3 = fig.add_subplot(gs[1, :])
ax3.plot(x_values, y_values_3, label='$y_3(x) = x \cdot \sin^2(10x)$', color= 'red')
ax3.set_title('График функции $y_3(x)$')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.legend()
ax3.grid(True)

plt.tight_layout()
plt.show()
```
